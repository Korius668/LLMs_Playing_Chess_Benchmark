## PHI model

```python
import pandas as pd
import os

df = pd.read_csv("../results/phi_model_results.csv")
```

```python
from matplotlib import pyplot as plt


phi_gain = df["Gain"]
mean_gain = phi_gain.mean()
plt.hist(phi_gain, bins=100, edgecolor="black")
plt.vlines(x=0, ymin=0, ymax=plt.gca().get_ylim()[1], colors="red", linestyles="dashed", label="Zero Gain")
plt.vlines(x=mean_gain, ymin=0, ymax=plt.gca().get_ylim()[1], colors="green", linestyles="solid", label=f"Mean Gain: {mean_gain:.2f}")
plt.xlabel("Gain")
plt.ylabel("Frequency")
plt.legend()
plt.title("Distribution of Phi Model Gains")
plt.show()
```

```python
import pandas as pd
import os
from_center = 0.1
df = pd.read_csv("../results/tournaments/tournament_16/match_R1_M1.csv")
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df["move_number"], df["centipawn_score"], label="centipawn_score", color="blue")
ax.axhline(y=0, color="red", linestyle="dashed", label="Even Position")
minimum = df["centipawn_score"].min()
maximum = df["centipawn_score"].max()
ax.text(
    1, from_center*maximum, "White PHI is better",
    transform=ax.get_yaxis_transform(),
    ha="right", va="bottom",
    fontsize=20, color="gray", alpha=0.9
)
ax.scatter(df["move_number"].iloc[-1], df["centipawn_score"].iloc[-1],marker="*",color="yellow", s=200, edgecolor="black")

# --- TEXT BELOW THE LINE ---
ax.text(
    1, from_center*minimum, "Black MIXTRAL is better",
    transform=ax.get_yaxis_transform(),
    ha="right", va="top",
    fontsize=20, color="black"
)
plt.xlabel("Move Number")
plt.ylabel("Centipawn Score")
plt.title("Centipawn Score over time match T1, R1, M1")
plt.legend()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("../results/tournaments/tournament_17/match_R1_M1.csv")
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df["move_number"], df["centipawn_score"], label="centipawn_score", color="blue")
ax.axhline(y=0, color="red", linestyle="dashed", label="Even Position")
minimum = df["centipawn_score"].min()
maximum = df["centipawn_score"].max()
ax.text(
    0.5, 0.5*1000, "White LLAMA-CHESS is better",
    transform=ax.get_yaxis_transform(),
    ha="center", va="bottom",
    fontsize=20, color="gray", alpha=0.9
)
ax.scatter(df["move_number"].iloc[-1], df["centipawn_score"].iloc[-1],marker="*",color="yellow", s=200, edgecolor="black")
# --- TEXT BELOW THE LINE ---
ax.text(
    0.5, -0.5*1000, "Black RANDOM is better",
    transform=ax.get_yaxis_transform(),
    ha="center", va="top",
    fontsize=20, color="black"
)
plt.xlabel("Move Number")
plt.ylabel("Centipawn Score")
plt.title("Centipawn Score over time match T2, R1, M1")
plt.legend()
plt.ylim([-1000,1000])
plt.show()
```
